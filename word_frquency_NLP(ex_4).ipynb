```python
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import urllib.request
from matplotlib import pyplot as plt

# Download the necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Retrieve a copy of the text from the URL
text_file = urllib.request.urlopen("https://www.gutenberg.org/cache/epub/64317/pg64317.txt")
text = text_file.read().decode('utf-8')  # Read and decode the text
# Preview a section of the text
print(text[0:100])   #enter the starting and ending points
```

```python
# Tokenize text by words
words = word_tokenize(text)

# Check the number of words
print(f"The total number of words in the text is {len(words)}")
```

```python
# Create an empty list to store words without punctuation and stopwords
words_no_punc = []

# Get the list of English stopwords
stop_words = set(stopwords.words('english'))

# Iterate through the words list to remove punctuations, stopwords, and convert to lowercase
for word in words:
    if word.isalpha() and word.lower() not in stop_words:
        words_no_punc.append(word.lower())

print(f"The total number of words without punctuation and stopwords is {len(words_no_punc)}")
```

```python
# Find the frequency of words
fdist = FreqDist(words_no_punc)

# Plot the 30 most common words
fdist.plot(30)
plt.show()
```
